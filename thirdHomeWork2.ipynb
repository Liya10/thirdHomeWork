```python
#первое задание
import matplotlib.pyplot as plt
import numpy as np
print("Первое задание")
data = np.loadtxt('populations.txt')
year, hare, lynx, carrot = data.T
plt.axes([0, 0, 0.8, 1]) 
plt.plot(year, hare,year, lynx, year, carrot) 
plt.legend(('Hare', 'Lynx', 'Carrot'), loc=(1.04, 0.4))
plt.xlabel(u'Годы')
plt.ylabel(u'Численность')
plt.show()
```

```python
#второе задание
import pandas as pd
print("Второе задание")
df=pd.DataFrame({'hare':hare,'lynx':lynx})
totals=[i+j for i,j in zip(df['hare'],df['lynx'])]
hareBars=[i/j *100 for i,j in zip(df['hare'],totals)]
lynxBars=[i/j *100 for i,j in zip(df['lynx'],totals)]
plt.bar(year,hareBars, color='#b5ffb9',edgecolor='white',width=1)
plt.bar(year,lynxBars,bottom=hareBars, color='#f9bc86',)
plt.legend(('кролики', 'рыси'), loc=(1.04, 0.4))
plt.show()

```

```python
#третье задание
print("Третье задание")
lynxMean=int(lynx.mean(axis=0))
hareMean=int(hare.mean(axis=0))
carrotMean=int(carrot.mean(axis=0))
print("Средняя численность рыси ", lynxMean,"и ее отклонение ",int(lynx.std(axis=0)))
plt.bar(year,lynx,color='#ffcc00')
plt.plot(year,np.repeat(lynxMean,21), 'r-', linewidth=2.5) 
plt.legend(( 'средняя численность рысей','рысь'), loc=(1.04, 0.4))
plt.show()
print("Средняя численность кроликов ", hareMean,"и ее отклонение ",int(hare.std(axis=0)))
plt.bar(year,hare,color='#99ccff')
plt.plot(year,np.repeat(hareMean,21), 'r-', linewidth=2.5) 
plt.legend(( 'средняя численность кроликов','кролик'), loc=(1.04, 0.4))
plt.show()
print("Средняя численность моркови ", carrotMean,"и ее отклонение ",int(carrot.std(axis=0)))
plt.bar(year,carrot,color='#ff944d')
plt.plot(year,np.repeat(carrotMean,21),'g-', linewidth=2.5 ) 
plt.legend(( 'средняя численность моркови','морковь'), loc=(1.04, 0.4))
plt.show()

```

```python
#четвертое задание
print()
print("Четвертое задание")
yearHare=np.argmax(hare)
yearLynx=np.argmax(lynx)
yearCarrot=np.argmax(carrot)
print("В ",int(year[yearHare]),"году кролики имели наибольшую численности")
print("В ",int(year[yearLynx]),"году рыси имели наибольшую численности")
print("В ",int(year[yearCarrot]),"году морковки имели наибольшую численности")
plt.plot(year, hare,year, lynx, year, carrot) 
plt.scatter([year[yearHare],year[yearLynx],year[yearCarrot]],[hare[yearHare],lynx[yearLynx],carrot[yearCarrot]], color='r')
plt.legend(('Кролик', 'Рысь', 'Морковь','Максимумы'), loc=(1.04, 0.4))
plt.show()

```

```python

#пятое задание
print()
print("Пятое задание")
S=np.repeat('H',21)
K=np.argmax(np.array((hare,lynx)),0)
S[K==1]='L'
print(S)
labels = ['Hare','Lynx']
k=np.sum(K)
x=[21-k,k]
explode = [0, 0.05]
plt.pie(x, labels = labels,explode = explode,  autopct = '%1.1f%%', shadow=True);
plt.show()
```

```python

# шестое задание
#уточнить задание
print("Шестое задание")
hareAbove50e3=hare[hare>50e3]
yearHareAbove50e3=year[hare>50e3]
lynxAbove50e3=lynx[lynx>50e3]
yearLynxAbove50e3=year[lynx>50e3]
carrotAbove50e3=carrot[carrot>50e3]
yearCarrotAbove50e3=year[carrot>50e3]
print("Года, в которых численность кроликов больше 50000 : ", yearHareAbove50e3)
print("Года, в которых численность рысей больше 50000 : ", yearLynxAbove50e3)
print("Года, в которых численность моркови больше 50000 : ", yearCarrotAbove50e3)
plt.plot(year, hare,year, lynx, year, carrot) 
plt.scatter(yearHareAbove50e3,hareAbove50e3)
plt.scatter(yearLynxAbove50e3,lynxAbove50e3)
plt.scatter(yearCarrotAbove50e3,carrotAbove50e3)
plt.legend(('Кролик', 'Рысь', 'Морковь','Кроликов более 50000','Рысей более 50000','Моркови более 50000'), loc=(1.04, 0.4))
plt.show()
```

```python
# седьмое задание
print("Седьмое задание")
yearHare=np.argsort(hare)
yearLynx=np.argsort(lynx)
yearCarrot=np.argsort(carrot)
print("В ",int(year[yearHare[0]])," и ", int(year[yearHare[1]]), "были наилучшими годами для кроликов")
print("В ",int(year[yearLynx[0]])," и ", int(year[yearLynx[1]]), "были наилучшими годами для рысей")
print("В ",int(year[yearCarrot[0]])," и ", int(year[yearCarrot[1]]), "были наилучшими годами для моркови")
```

```python
#восьмое задание
print("Восьмое задание")
hare_grad=np.gradient(hare,1.0)
plt.plot(year,hare_grad,year, lynx)
plt.legend(('Изменение численности кроликов', 'Число рысей'), loc=(1.04, 0.4))
plt.xlabel(u'Годы')
plt.ylabel(u'Численность')
plt.show()
print("коэффициент корреляции ",np.corrcoef(hare_grad,lynx)[0,1])
```

```python
#девятое задание
print("Девятое задание")
plt.scatter(hare, carrot, color="#000066")
plt.xlabel(u'Кролики')
plt.ylabel(u'Моркови')
plt.show()
plt.scatter(hare, lynx, color="#000066")
plt.xlabel(u'Кролики')
plt.ylabel(u'Рысей')
plt.show()
```

```python
#десятое задание
import pandas as pd
import cmath 
print("Десятое задание")
maxHare=np.max(hare)
minHare=np.min(hare)
maxHare_grad=np.max(hare_grad)
minHare_grad=np.min(hare_grad)
maxLynx=np.max(lynx)
maxCarrot=np.max(carrot)
print("График изменения численности кроликов за определенные года")
plt.plot(year,hare_grad)
plt.show()
print("График синуса, который приближен к предыдущему графику ")
arc=cmath.asin((hare_grad[0]-minHare_grad)/(maxHare_grad-minHare_grad))*np.pi
x=np.linspace(-np.pi+arc, 3*np.pi+arc, 21, endpoint=True)
S=np.sin(x)*((maxHare_grad-minHare_grad)/2)+(maxHare_grad+minHare_grad)/2
plt.plot(x,S)
plt.show()

hare_fut=int(hare[19]+S[20])
print("Прогнозирование численности кроликов за 1920 год: ",hare_fut)#прогнозирование кроликов на 1920
print("Отклонение от действительности : ",np.abs(hare_fut-int(hare[20])))#ошибка при прогнозе для кроликов
#maxLynx=a*minHare+b
a=np.corrcoef(hare_grad,lynx)[0,1]
b=maxLynx-a*minHare
lynx_fut=int(a*hare_fut+b)

print("Прогнозирование численности рысей: ",lynx_fut)
print("Отклонение от действительности : ",np.abs(lynx_fut-lynx[20]))

a=np.corrcoef(hare_grad,carrot)[0,1]
b=maxCarrot-a*minHare
carrot_fut=int(a*hare_fut+b)
# у моркови 
print("Прогнозирование численности моркови : ",carrot_fut)
print("Отклонение от действительности : ",np.abs(carrot_fut-carrot[20]))
print("У моркови серьезное отклонение, потому слабее зависит от численности кроликов и более того, если посмотреть на график, который распологается чуть, то сложно определить тенденцию изменения численности моркови.")
print("График изменения численности моркови за определенные года")
carrot_grad=np.gradient(carrot,1.0)
maxCarrot=np.max(carrot)
minCarrot=np.min(carrot)
maxCarrot_grad=np.max(carrot_grad)
minCarrot_grad=np.min(carrot_grad)

plt.plot(year,carrot_grad)
plt.show()


```
